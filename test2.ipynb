```python
print("a", " ".join("{:3}".format(i) for i in a))
print("b", " ".join("{:3}".format(i) for i in b))
print("c", " ".join("{:3}".format(i) for i in c))
print("d", " ".join("{:3}".format(i) for i in d))
print("e", " ".join("{:3}".format(i) for i in e))
```

```python
# Find intersections between b and others
set_b = set(b)
inter_b_a = set_b & set(a)
inter_b_c = set_b & set(c)
inter_b_d = set_b & set(d)
inter_b_e = set_b & set(e)

print("b ∩ a:", sorted(inter_b_a))
print("b ∩ c:", sorted(inter_b_c))
print("b ∩ d:", sorted(inter_b_d))
print("b ∩ e:", sorted(inter_b_e))
```

```python
# Find intersections between c and others
set_c = set(c)
inter_c_a = set_c & set(a)
inter_c_b = set_c & set(b)
inter_c_d = set_c & set(d)
inter_c_e = set_c & set(e)

print("c ∩ a:", sorted(inter_c_a))
print("c ∩ b:", sorted(inter_c_b))
print("c ∩ d:", sorted(inter_c_d))
print("c ∩ e:", sorted(inter_c_e))
```
